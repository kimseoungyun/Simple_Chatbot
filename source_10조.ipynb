```python
# Building dictionary of Intents & Keywords

positive = []
negative = []
invert = []
patty = []
meatTopping = []
vegetableTopping = []
sauce = []
vegetable = []
hi = []
bye = []
plz = []
all = []

keywords_dict={}
for key in p_list_syn.keys():
    for synonym in list(p_list_syn[key]):
        positive.append(re.compile('\\b'+synonym+'\\b'))

keywords_dict["positive"] = positive

for key in n_list_syn.keys():
    for synonym in list(n_list_syn[key]):
        negative.append(re.compile('\\b'+synonym+'\\b'))
        
keywords_dict["negative"] = negative        
        
for key in not_list_syn.keys():
    for synonym in list(not_list_syn[key]):
        invert.append(re.compile('\\b'+synonym+'\\b'))
        
keywords_dict["not"] = invert

for key in v_list_syn.keys():
    for synonym in list(v_list_syn[key]):
        vegetable.append(re.compile('\\b'+synonym+'\\b'))
        
keywords_dict["vegetable"] = vegetable

for key in hi_list_syn.keys():
    for synonym in list(hi_list_syn[key]):
        hi.append(re.compile('\\b'+synonym+'\\b'))
        
keywords_dict["hi"] = hi

for key in bye_list_syn.keys():
    for synonym in list(bye_list_syn[key]):
        bye.append(re.compile('\\b'+synonym+'\\b'))
        
keywords_dict["bye"] = bye

for key in plz_list_syn.keys():
    for synonym in list(plz_list_syn[key]):
        plz.append(re.compile('\\b'+synonym+'\\b'))
        
keywords_dict["plz"] = plz

for key in all_list_syn.keys():
    for synonym in list(all_list_syn[key]):
        all.append(re.compile('\\b'+synonym+'\\b'))
        
keywords_dict["all"] = all

count = 0
for key in i_list_syn.keys():
    if count <5:
        for synonym in list(i_list_syn[key]):
            patty.append(re.compile('\\b'+synonym+'\\b'))
    elif count <7:
        for synonym in list(i_list_syn[key]):
            meatTopping.append(re.compile('\\b'+synonym+'\\b'))
    elif count <11:
        for synonym in list(i_list_syn[key]):
            vegetableTopping.append(re.compile('\\b'+synonym+'\\b'))
    elif count <17:
        for synonym in list(i_list_syn[key]):
            sauce.append(re.compile('\\b'+synonym+'\\b'))
    else:
        for synonym in list(i_list_syn[key]):
            meatTopping.append(re.compile('\\b'+synonym+'\\b'))
            
    count += 1
        
keywords_dict["patty"] = patty
keywords_dict["meatTopping"] = meatTopping
keywords_dict["vegetableTopping"] = vegetableTopping
keywords_dict["sauce"] = sauce

keywords_dict
```

```python
# 주어진 딕셔너리와 문자열 이용 단어 추출하는 함수111
def search1(keywords_dict, user_input, l, r):
    matched = []
    for intent, pattern_list in keywords_dict.items():
        for pattern in pattern_list:
            info = pattern.search( user_input[l:r])
            if info != None: 
                matched.append((intent, info.span(), info.group()))

    matched.sort(key = lambda x:x[1][0])
    return matched
                                
```
